```python
import tkinter as tk

def get_age():
    try:
        age = int(age_entry.get())
    except ValueError:
        input_frame.pack_forget()
        error_frame.pack(fill="both", expand=True)
        error_label.config(text="Please enter numbers")
        back_button.pack(pady = 10) 
    except TypeError:
        input_frame.pack_forget()
        error_frame.pack(fill="both", expand=True)
        error_label.config(text="Please enter a number that is not zero")
        back_button.pack(pady = 10)  
    else:
        input_frame.pack_forget()
        name_frame.pack(fill="both", expand=True)
        name_label.pack(pady = 10)
        name_entry.pack(pady = 10)
        name_label.config(text="What's your name?")
        name_submit_button.pack(pady = 10)
        get_age.age = age
        
def go_back():
    error_frame.pack_forget()
    input_frame.pack(fill="both", expand=True)
    error_label.config(text="")
    
def get_name():
    try:
        name = str(name_entry.get())
        if name.isdigit():
            raise TypeError
        if name.isspace():
            raise TypeError
    except TypeError:
        name_frame.pack_forget()
        error_frame.pack(fill="both", expand=True)
        error_label.config(text="Please enter a valid name")
        back_buttonName.pack(pady = 10)
    else:
        name_frame.pack_forget()
        height_frame.pack(fill="both", expand=True)
        height_label.config(text="What's your height?")
        height_submit_button.pack(pady = 10)
        get_name.name = name
        
def go_backName():
    error_frame.pack_forget()
    name_frame.pack(fill="both", expand=True)
    error_label.config(text="")
    back_buttonName.pack_forget()
    
def get_height():
    try:
        height = str(height_entry.get())
        if "'" not in height:
            raise TypeError
        if height.endswith("'"):
            raise TypeError
        if " " in height:
            raise TypeError
        feet, inches = height.split("'")
        height_frame.pack_forget()
        get_height.height = int(feet)*12+int(inches)      
        result_frame.pack(fill="both", expand=True)
        result_label.config(text="Hello, " + str(get_name.name) + ". You are " + str(get_age.age) + ". You are " + str(get_height.height)+" inches tall.") 
    except TypeError:
        height_frame.pack_forget()
        error_frame.pack(fill="both", expand=True)
        error_label.config(text="Please enter a valid height in feet")
        back_buttonHeight.pack(pady = 10)
        
def go_backHeight():
    error_frame.pack_forget()
    height_frame.pack(fill="both", expand=True)
    error_label.config(text="")
    back_buttonHeight.pack_forget()
        
#
root = tk.Tk()
root.title("Age and Name Input")
root.geometry("600x200")

input_frame = tk.Frame(root, bg="blue")
input_frame.pack(fill="both", expand=True) 

age_label = tk.Label(input_frame, text="How old are you currently?", fg="white", font=("Arial", 12), bg="blue")
age_label.pack(pady = 10)

age_entry = tk.Entry(input_frame, font=("Arial", 12))
age_entry.pack(pady = 10)

submit_button = tk.Button(input_frame, text="Submit", command=get_age, font=("Arial", 12), bg="lightblue")
submit_button.pack(pady = 10)

error_frame = tk.Frame(root, bg="blue")

error_label = tk.Label(error_frame, text="", fg="red", font=("Arial", 12), bg="blue")
error_label.pack(pady = 10)

back_button = tk.Button(error_frame, text="Go Back", command=go_back, font=("Arial", 12), bg="lightblue")

#

name_frame = tk.Frame(root, bg="blue")

name_label = tk.Label(name_frame, text="", fg="white", font=("Arial", 12), bg="blue")

name_entry = tk.Entry(name_frame, font=("Arial", 12))

name_submit_button = tk.Button(name_frame, text="Submit", command=get_name, font=("Arial", 12), bg="lightblue")

back_buttonName = tk.Button(error_frame, text="Go Back", command=go_backName, font=("Arial", 12), bg="lightblue")

#

height_frame = tk.Frame(root, bg="blue")

height_label = tk.Label(height_frame, text="", fg="white", font=("Arial", 12), bg="blue")
height_label.pack(pady = 10)

height_entry = tk.Entry(height_frame, font=("Arial", 12))
height_entry.pack(pady = 10)

height_submit_button = tk.Button(height_frame, text="Submit", command=get_height, font=("Arial", 12), bg="lightblue")

back_buttonHeight = tk.Button(error_frame, text="Go Back", command=go_backHeight, font=("Arial", 12), bg="lightblue")

#

result_frame = tk.Frame(root, bg="blue")

result_label = tk.Label(result_frame, text="", fg="white", font=("Arial", 14), bg="blue")
result_label.pack(fill="both", expand=True)

root.mainloop()
```
